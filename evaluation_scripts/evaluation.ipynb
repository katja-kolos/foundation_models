```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("./all_metrics_val.xlsx")
```

```python
plt.figure(figsize=(16, 12))
ax = sns.barplot(
    x='setting', 
    y='acc_average', 
    hue='model', 
    data=df[df["acc_language"].notna()], 
    ci=None, 
    palette="coolwarm"
)

for container in ax.containers:
    ax.bar_label(container, fmt='%.1f', fontsize=10, padding=3)

plt.title('Average Accuracy by Model and Setting', fontsize=20)
plt.ylabel('Average Accuracy (%)', fontsize=20)
plt.xlabel('Settings', fontsize=20)
plt.legend(title='Model', fontsize=10, loc='lower left', )
plt.ylim(0, 100)
plt.savefig(f"./images/acc_average.pdf")
plt.show()
```

```python
plt.figure(figsize=(16, 12))
ax = sns.barplot(x='setting', y='overall_metric', data=df, hue='model', ci=None, palette="coolwarm")

for container in ax.containers:
    ax.bar_label(container, fmt='%.1f', fontsize=12, padding=3)

plt.title('Evaluation of explanations by Model and Setting', fontsize=25)
plt.ylabel('Overall Metric (%)', fontsize=20)
plt.xlabel('Settings', fontsize=20)
plt.legend(title='Model', fontsize=17, loc="lower left")
plt.ylim(0, 40)
plt.savefig(f"./images/overall_metrics.pdf")
plt.show()
```
